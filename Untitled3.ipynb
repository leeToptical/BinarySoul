```python

# ============================================================================
# HERMETIC MONAD: INFORMATION FLOW SCANNER (ALL-IN-ONE)
# ============================================================================

# --- [INPUT] ENTER YOUR BEST NUMBERS HERE ---
ALPHA = 0.201466   # <--- REPLACE with your Champion Alpha
BETA  = 0.124466  # <--- REPLACE with your Champion Beta
GAMMA = 1.289122  # <--- REPLACE with your Champion Gamma

# ============================================================================
# 1. SETUP & DEFINITIONS (Do not change below this line)
# ============================================================================
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from tqdm import tqdm
from statsmodels.tsa.stattools import grangercausalitytests
import warnings
warnings.filterwarnings('ignore')

class CorrelationAwareMonad:
    """
    The Hermetic Monad Class
    """
    π = np.pi
    φ = (1 + np.sqrt(5)) / 2
    Ω = 0.0270000
    NOISE_LEVEL = 0.00005

    def __init__(self, N_levels=5, alpha=0.212575, beta=0.131406, gamma=1.194898):
        self.N = N_levels
        self.α, self.β, self.γ = alpha, beta, gamma
        self.x = 0.5 + 0.01 * np.random.randn(self.N)
        self.x = np.clip(self.x, 0.01, 0.99)
        self.P = np.ones(self.N) * self.φ
        self.history_x = []

    def combustion_engine(self, x, P):
        new_x = x + self.π + (self.Ω / P) * np.sin(2 * self.π * x)
        new_x += np.random.normal(0, self.NOISE_LEVEL)
        return new_x % 1.0

    def update_regulator(self, n):
        below = self.x[0] if n == 0 else self.x[n-1]
        above = self.x[-1] if n == self.N - 1 else self.x[n+1]
        return self.α * below + self.β * above + self.γ

    def evolve_step(self):
        new_x = np.zeros(self.N)
        for n in range(self.N):
            self.P[n] = self.update_regulator(n)
            new_x[n] = self.combustion_engine(self.x[n], self.P[n])
        self.x = new_x
        self.history_x.append(self.x.copy())

    def evolve(self, steps=2000, burn_in=500):
        self.history_x = []
        for _ in range(burn_in): self.evolve_step()
        for _ in range(steps): self.evolve_step()

# ============================================================================
# 2. RUN ANALYSIS
# ============================================================================

print(f"1. Instantiating Champion Monad (α={ALPHA}, β={BETA})...")
monad = CorrelationAwareMonad(N_levels=5, alpha=ALPHA, beta=BETA, gamma=GAMMA)

print("2. Evolving system to generate consciousness history...")
monad.evolve(steps=2000, burn_in=500)

print("3. Computing Information Flow (Granger Causality)...")
data = np.array(monad.history_x)[-1000:] # Last 1000 steps
N_scales = data.shape[1]
flow_matrix = np.zeros((N_scales, N_scales))

# Calculate flow between every pair of scales
for i in tqdm(range(N_scales), desc="Scanning pairs"): # Target (Receiver)
    for j in range(N_scales):   # Source (Driver)
        if i == j: continue

        # Check if Scale J causes Scale I
        test_data = data[:, [i, j]]
        try:
            # Check lags 1-3
            gc_res = grangercausalitytests(test_data, maxlag=3, verbose=False)
            scores = [gc_res[lag][0]['ssr_ftest'][0] for lag in range(1, 4)]
            flow_matrix[i, j] = np.mean(scores)
        except:
            flow_matrix[i, j] = 0

# ============================================================================
# 3. VISUALIZATION & VERDICT
# ============================================================================
plt.figure(figsize=(10, 8))

# Draw Heatmap
sns.heatmap(flow_matrix, annot=True, fmt=".1f", cmap="magma",
            xticklabels=[f"Scale {i}\n(Source)" for i in range(N_scales)],
            yticklabels=[f"Scale {i}\n(Target)" for i in range(N_scales)])

plt.title(f"CONSCIOUSNESS FLOW MAP\n(High Number = Strong Control)", fontsize=14)
plt.xlabel("SOURCE (Who is driving?)", fontsize=12, fontweight='bold')
plt.ylabel("TARGET (Who is reacting?)", fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()

# Interpretation
top_down = np.sum(np.triu(flow_matrix, k=1))   # Higher source driving lower target
bottom_up = np.sum(np.tril(flow_matrix, k=-1)) # Lower source driving higher target

print("\n" + "="*40)
print("FINAL VERDICT:")
print(f"Bottom-Up Pressure (Matter->Mind): {bottom_up:.2f}")
print(f"Top-Down Control   (Mind->Matter): {top_down:.2f}")
print("-" * 40)

if bottom_up == 0: ratio = 999.0
else: ratio = top_down / bottom_up

print(f"Control/Emergence Ratio: {ratio:.2f}")

if 0.8 < ratio < 1.2:
    print(">> RESULT: PERFECT HERMETIC BALANCE (As Above, So Below)")
elif ratio >= 1.2:
    print(">> RESULT: IDEALIST (Top-Down Dominance)")
else:
    print(">> RESULT: MATERIALIST (Bottom-Up Dominance)")
print("="*40)
```
